```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
cols = ['id','gender','age','income','score']
df = pd.read_csv('Mall_Customers.csv',names=cols,header=0)
df.head(5)
```

```python
x = 'income'
y = 'score'
X = df[[x,y]].values
```

```python
for i in range (len(cols)-1):
  for j in range (i+1,len(cols)-1):
    x1 = cols[i]
    y1 = cols[j]
    sns.scatterplot(x=x1,y=y1,data=df,hue = 'income', palette='viridis', s=100, alpha=0.8)
    plt.show()
```

```python
model = KMeans(n_clusters=5)
model.fit(X)
```

```python
y_pred = model.predict(X)
```

```python
y_pred
```

```python
sns.scatterplot(data = df,x=x1,y=y1,hue = y_pred,palette='viridis', s=100, alpha=0.8)
```

```python
plt.figure(figsize=(10, 6))
df['Cluster'] = model.predict(X)


sns.scatterplot(data=df, x=x1, y=y1,hue=df['Cluster'], palette='viridis', s=100, alpha=0.8)
plt.xlabel('Annual Income ($1000s)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segmentation using K-Means')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(X, model.labels_)
print(f"Silhouette Score: {silhouette_avg}")
```

```python
inertia = model.inertia_
print(f"Inertia: {inertia}")
```
